#### Through an analysis with Tableau, I saw that in Europe, the largest concentration of companies is in the UK around London, so I took the latitude and longitude of London as a reference to do my search in a radius of 100 km around. 

#### The latitude and longitude I took as reference were: 

### "51.507359", "-0.136439"


```python
from dotenv import load_dotenv
import requests
from urllib.parse import urlencode
import os
import pandas as pd
```

```python
load_dotenv()
```

```python
api_key = os.getenv("patata")
```

```python
def place_location(lat, lng, keyword):
    
    places = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    parameters = {
        "key":api_key,
        "location":f"{lat},{lng}",
        "radius":100000,
        "keyword":f"{keyword}"
    }
    parameters_encoded = urlencode(parameters)
    url = f"{places}?{parameters_encoded}"

    result = requests.get(url)
    return result   
```

```python
### Starbucks

starbucks = place_location("51.507359", "-0.136439", "Starbucks")
```

```python
df = pd.DataFrame(columns = ["type", "name", "lat", "lng"])

locs_dict = {
    "Vegan": veg_location,
    "Dog groomer": doggrom_location,
    "Pubs": pub_location,
    "Schools": school_location,
    "Airport": airport_location,
    "Basket": basket_stadium,
    "Design": design_location,
    "Starbucks": starbucks_location
    }

index = 0
for loctype, loclist in locs_dict.items():
    for place in loclist:
        df.loc[index] = {
            "type": loctype,
            "name": place[0],
            "lat": place[1]["lat"],
            "lng": place[1]["lng"]
        }
        index += 1

df.head()
```
